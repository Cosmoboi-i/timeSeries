```python
import pandas as pd
import numpy as np
df = pd.read_excel('DATASET.xlsx')
df
```

```python
df.head(30)
```

```python
df_test = df.loc[df['year'] == 10]
df_test
```

```python
df = df.drop(df[df['year'] == 10].index)
```

```python
df.head(12)
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df_test = df_test.drop(['Unnamed: 0'], axis=1)
```

```python
df_test
```

#Model for Para-9

```python
train = df.iloc[:, :10]
train
```

```python
from pycaret.regression import *
s = setup(data=train, target='Para-9', train_size = 0.95, silent=True)
cm = compare_models()
```
